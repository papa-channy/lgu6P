## loc vs iloc

```python
# tips.loc[tips['day']=="Sat"]
# tips.loc[행, 열]
tips.loc[0:2,['total_bill','tip','day']]
```

```python
tips.iloc[0:4,[0, 1, 4]]
```

```python
tips.loc[tips['day']=="Sat",['total_bill','tip','day']].reset_index(drop=1)
```

```python
tips.loc[:,:]
```

```python

tips.loc[행 조건식만 만들자, :]
```

```python
# total_bill 11이하인것만 조회
tips.loc[tips['total_bill']<=11,:].head(1)
```

```python
#time이 dinner인 것만 조회, loc적용
tips.loc[tips['time']=='Dinner',:]
```

```python
# time이 dinner이면서 total_bill이 11이하인 것만 조회
a = tips.loc[tips['time']=='Dinner',:]
a.loc[a['total_bill']<=11,:].head(2)
```

```python
# 조건식 2개 이상
# 필터링 문법: tips.loc[(조건식) $ (조건식) $(조건식), : ]
tips.loc[(tips['time']=='dinner') &tips['total_bill']<=11,:].reset_index(drop=1)
```
